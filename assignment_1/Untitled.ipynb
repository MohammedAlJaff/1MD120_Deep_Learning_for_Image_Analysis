```python
from maj_linear_model import LinearRegressionModel
import numpy as np
import matplotlib.pyplot as plt
from load_auto import load_auto
```

```python
def standarize_data(X):
    'Assume X is a data matrix with n rows/data points, '
    
    feature_means = np.mean(X, axis=0)
    feature_std = np.std(X, axis=0)
    
    standardized_X = (X - feature_means)/feature_std
    return standardized_X
```

```python
# Load automobile data-set 
Xraw, y = load_auto()
Xraw
```

```python
# Standardize data matrix
X = standarize_data(Xraw)

horsepower_column_j = 2
X7 = X
X1 = X[:,horsepower_column_j].reshape([np.size(X[:,horsepower_column_j]),1])
```

```python
# Instansiate Linear regression model
maj_7_model = LinearRegressionModel(data_X=X, true_label_Y=y)
w7, b7, j7 =  maj_7_model.train_linear_model(X=X, y_true=y, nr_iter=400)

print('......')
print('Model 2 - All features case: ')
print('weights w: ', w7)
print('offset w: ', b7)
```

```python
maj_1_model = LinearRegressionModel(data_X=X1, true_label_Y=y)
w1, b1, j1 =  maj_1_model.train_linear_model(X=X1, y_true=y, nr_iter=1000)
print('......')
print('Model 1 - Horsepower feature only case: ')
print('weights w: ', w1)
print('offset w: ', b1)
```

```python
learning_rates = [1, 1e-1, 1e-2, 1e-3, 1e-4]
training_curves = []

for i in range(len(learning_rates)):
    
    #define model 
    maj_1_model = LinearRegressionModel(data_X=X1, true_label_Y=y)
    
    # fit model with learning rate 
    lr = learning_rates[i]
    w1, b1, j1 =  maj_1_model.train_linear_model(X=X1, y_true=y, nr_iter=1000, learning_rate=lr)
    training_curves.append(j1)
    # same and append trainng cost trajectories

    
plt.figure(figsize=[15,5])
for i in range(len(learning_rates)):
    plt.plot(training_curves[i], label='$\\alpha$ = '+str(learning_rates[i]))
    plt.legend()

plt.legend()
plt.xlabel('iterations')
plt.ylabel('Training Cost/Emperical Risk')
plt.title('Model 1: Using only  "Horsepower" as the input data feature')
```
